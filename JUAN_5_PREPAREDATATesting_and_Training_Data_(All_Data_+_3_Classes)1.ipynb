# **# 1 - Case Folding**
Pembersihan huruf dari huruf kapital menjadi huruf kecil dan juga penghilangan angka serta tanda baca yang ada.
1. Remove numbers from string
2. Change strings to lowercase
3. Remove punctuations (annotation removal)
4. Remove repeated characters (duplicate) characters from a string
5. Remove emojis

```python
def convertToString(term):
  if type(term) is str:
    return term
  else:
    return str(term)
data_utube['comment'] = data_utube['comment'].apply(convertToString)
```

```python
#PROSES CLEANING DATA

# =====================================================================================
# 1. Remove numbers from string
# =====================================================================================
data_utube['comment'] = data_utube['comment'].str.replace('\d+', '', regex=True)

# =====================================================================================
# 2. Change Strings to lowercase in Pandas DataFrame
# =====================================================================================
data_utube['comment'] = data_utube['comment'].apply(lambda x:x.lower())


# =====================================================================================
# 3. Remove punctuations (Annotation Removal)
# =====================================================================================
print(string.punctuation);

# Function to remove punctuations
def remove_punctuations(text):
  for sym_punc in string.punctuation:
    text = text.replace(sym_punc, ' ')
    text = re.sub(' +', ' ', text) # replace multiple spaces with single space
  return text;

data_utube["comment"] = data_utube['comment'].apply(remove_punctuations)


# =====================================================================================
# 4. Remove repeated characters (duplicate) characters from a string
# =====================================================================================
# Function to remove duplcate characters
def remove_duplcate_chars(text):
  pattern = r'(.)\1+' # (.) any character repeated (\+) more than
  repl = r'\1'        # replace it once
  text = re.sub(pattern, repl, text);
  return text;

data_utube["comment"] = data_utube['comment'].apply(remove_duplcate_chars)

# =====================================================================================
# 5. Remove emojis
# =====================================================================================
# Remove emojis (sudah di cek dan hasilnya semua emoji bisa di hapus)
data_utube['comment'] = data_utube['comment'].apply(lambda x: x.encode('ascii', 'ignore').decode('ascii'))

data_utube
```
